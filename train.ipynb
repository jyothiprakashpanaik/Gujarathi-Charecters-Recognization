```
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import time 

(data,labels,names,idx) = ([],[],{},0)

dataset = '/content/drive/My Drive/ML and DL pro/Gujarathi/train (1)'
# print(list(os.listdir(dataset)))

for subdirs, dirs, files in os.walk(dataset):
    for subdir in dirs:
        names[idx] = subdir
        path = os.path.join(dataset,subdir)
        for file in os.listdir(path):
            try:
                image = Image.open(os.path.join(path,file))
                image = image.resize((30,30))
                image = np.array(image)
                data.append(image)
                labels.append(idx)
                num = str(idx)
                print("{0} - {1} - {2}".format(num.zfill(5), subdir, file))
            except Exception as e:
                print(e)
        idx += 1
print('Dataset Loaded')
```

```
# Conver list to array
data = np.array(data)
labels = np.array(labels)

print(names)
print(data.shape, labels.shape)

# Spliting in train_test_split of the dataset
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# Converting the labels into on hot label encoding
y_train = to_categorical(y_train,34)
y_test = to_categorical(y_test,34)

# Bulding model 
print("""
	Building model
	""")
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu',input_shape=X_train.shape[1:]))
model.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(34,activation='softmax'))
print("Initialize model")
model.summary()

# Compilation of the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train,y_train,batch_size=100,epochs=250,validation_steps=(X_test,y_test))
```

```
model_json = model.to_json()
with open('model.json','w') as json_file:
	json_file.write(model_json)
model.save('model.h5')
```

```
from keras.models import model_from_json,load_model
import numpy as np
from keras.preprocessing import image
from PIL import Image
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

dataset = '/content/drive/My Drive/ML and DL pro/Gujarathi/test'
model = load_model('model.h5')

# img = mpimg.imread(path)
# imgplot = plt.imshow(img)

for (subdir,dirs,files) in os.walk(dataset):
    for subdir in dirs:
        path = os.path.join(dataset,subdir)
        correct = 0

        for file in os.listdir(path):
            path1 = os.path.join(path,file)
            test_image = Image.open(path1)
            test_image = test_image.resize((30,30))
            test_image = np.expand_dims(test_image, axis=0)
            test_image = np.array(test_image)

            result = model.predict_classes(test_image)[0]
            sign = names[result]
            if sign == subdir:
                correct += 1
        total = len(os.listdir(path))
        percent = correct * 100//total
        print("correct - {0} / {1} . sucess percent {2} for {3}".format(str(correct), str(total), str(percent), subdir))
```
